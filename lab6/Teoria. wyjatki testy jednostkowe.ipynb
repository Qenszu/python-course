# Laboratorium 6. Wyjątki i testy jednostkowe w Pythonie




## 1. Wprowadzenie

Podczas działania programu mogą pojawiać się różnego rodzaju błędy, np.:

- niepoprawne dane wejściowe,
- brak pliku,
- dzielenie przez zero,
- próba wykonania niedozwolonej operacji.

Program powinien być odporny na takie sytuacje i reagować w kontrolowany sposób. W tym celu w Pythonie wykorzystuje się mechanizm **wyjątków** (*exceptions*).

Drugim ważnym elementem tworzenia oprogramowania są **testy jednostkowe**, które pozwalają automatycznie sprawdzać poprawność działania programu.


## 2. Wyjątki

Wyjątek to obiekt reprezentujący błąd wykryty podczas wykonywania programu. Gdy wystąpi wyjątek, normalne wykonywanie programu zostaje przerwane i interpreter szuka odpowiedniej obsługi błędu.

Przykładowe wyjątki wbudowane:

- `ValueError` — niepoprawna wartość,
- `TypeError` — niepoprawny typ danych,
- `FileNotFoundError` — brak pliku,
- `ZeroDivisionError` — dzielenie przez zero.

Dokumentacja:  
https://docs.python.org/3/tutorial/errors.html


```python
# Przykład wyjątku ValueError

value = "abc"

try:
    number = int(value)
except ValueError:
    print("Nie można zamienić tekstu na liczbę całkowitą.")
```

## 3. Obsługa wyjątków — `try` / `except`

Do obsługi błędów służy konstrukcja `try` / `except`.

Kod, który może spowodować błąd, umieszcza się w bloku `try`.  
Kod obsługujący błąd umieszcza się w bloku `except`.

Dzięki temu program nie musi kończyć działania po wystąpieniu błędu.


```python
try:
    number = int(input("Podaj liczbę: "))
    print("Podana liczba:", number)
except ValueError:
    print("To nie jest poprawna liczba całkowita.")
```

## 4. Zgłaszanie wyjątków — `raise`

Programista może samodzielnie zgłaszać wyjątki przy użyciu instrukcji `raise`.

Stosuje się ją wtedy, gdy program wykryje sytuację niepoprawną, np. ujemny wiek, pustą nazwę użytkownika albo próbę wypłaty większej kwoty niż dostępne saldo.


```python
def validate_age(age):
    if age < 0:
        raise ValueError("Wiek nie może być ujemny.")
    return True


try:
    validate_age(-5)
except ValueError as error:
    print("Błąd:", error)
```

## 5. Własne wyjątki

W większych projektach często tworzy się własne klasy wyjątków, aby dokładniej opisywać błędy.

Własny wyjątek tworzy się jako klasę dziedziczącą po `Exception`.

Takie podejście jest szczególnie przydatne wtedy, gdy w programie występuje kilka różnych typów błędów, np.:

- brak miejsc,
- ponowna rezerwacja,
- brak rezerwacji,
- błędna struktura pliku.


```python
class NoSeatsAvailableError(Exception):
    pass


class AlreadyReservedError(Exception):
    pass


def reserve(seats):
    if seats <= 0:
        raise NoSeatsAvailableError("Brak wolnych miejsc.")
    return seats - 1


try:
    reserve(0)
except NoSeatsAvailableError as error:
    print("Błąd rezerwacji:", error)
```

## 6. Blok `finally`

Blok `finally` wykonywany jest zawsze — niezależnie od tego, czy wystąpił wyjątek.

Najczęściej używany jest do:

- zamykania plików,
- zwalniania zasobów,
- sprzątania po operacjach.


```python
try:
    print("Próba wykonania operacji.")
    result = 10 / 2
    print("Wynik:", result)
except ZeroDivisionError:
    print("Nie można dzielić przez zero.")
finally:
    print("Ten blok wykona się zawsze.")
```

## 7. Walidacja danych

Walidacja polega na sprawdzaniu poprawności danych wejściowych przed ich użyciem.

Przykłady walidacji:

- sprawdzenie, czy imię nie jest puste,
- sprawdzenie, czy wiek jest liczbą całkowitą,
- sprawdzenie zakresu oceny,
- sprawdzenie, czy plik zawiera wymagane kolumny.

Walidacja często wykorzystuje:

- instrukcje warunkowe `if`,
- wyjątki `raise`,
- własne klasy wyjątków.


```python
def validate_user(name, age, email):
    if not name:
        raise ValueError("Imię nie może być puste.")

    if not isinstance(age, int):
        raise ValueError("Wiek musi być liczbą całkowitą.")

    if age < 18:
        raise ValueError("Użytkownik musi mieć co najmniej 18 lat.")

    if "@" not in email or "." not in email.split("@")[-1]:
        raise ValueError("Niepoprawny adres e-mail.")

    return True


validate_user("Anna", 25, "anna@example.com")
```

## 8. Testy jednostkowe

Testy jednostkowe służą do automatycznego sprawdzania działania małych fragmentów programu, np. pojedynczych funkcji lub metod.

Celem testów jest:

- wykrywanie błędów,
- sprawdzanie poprawności działania,
- zabezpieczenie programu przed regresją,
- ułatwienie rozwoju projektu.

Dobry test powinien:

1. przygotować dane,
2. wykonać funkcję lub metodę,
3. sprawdzić wynik.


## 9. Testy z użyciem `assert`

Najprostszy test można zapisać przy użyciu instrukcji `assert`.

Jeżeli warunek po `assert` jest fałszywy, test kończy się błędem.


```python
def add(a, b):
    return a + b


assert add(2, 3) == 5
assert add(-1, 1) == 0

print("Wszystkie testy zakończone powodzeniem.")
```

## 10. Biblioteka `unittest`

`unittest` to wbudowany framework testowy w Pythonie.

Testy zapisuje się w klasach dziedziczących po `unittest.TestCase`.

Dokumentacja:  
https://docs.python.org/3/library/unittest.html


```python
import unittest

def multiply(a, b):
    return a * b


class TestMultiply(unittest.TestCase):

    def test_multiply_positive_numbers(self):
        self.assertEqual(multiply(2, 3), 6)

    def test_multiply_by_zero(self):
        self.assertEqual(multiply(10, 0), 0)


suite = unittest.TestLoader().loadTestsFromTestCase(TestMultiply)
unittest.TextTestRunner(verbosity=2).run(suite)
```

## 12. Testowanie wyjątków w `pytest`

W testach należy sprawdzać nie tylko poprawne działanie programu, ale również poprawną obsługę błędów.

Do testowania wyjątków w `pytest` służy `pytest.raises()`.


```python
%%writefile test_validation.py
import pytest

def validate_age(age):
    if not isinstance(age, int):
        raise ValueError("Wiek musi być liczbą całkowitą.")
    if age < 18:
        raise ValueError("Wiek musi być większy lub równy 18.")
    return True


def test_valid_age():
    assert validate_age(20) is True


def test_age_too_low():
    with pytest.raises(ValueError):
        validate_age(15)


def test_invalid_age_type():
    with pytest.raises(ValueError):
        validate_age("20")
```

```python
!pytest -q test_validation.py
```

## 13. Testy parametryzowane

Testy parametryzowane pozwalają uruchomić ten sam test dla wielu zestawów danych.

Dzięki temu można uniknąć powielania kodu i łatwiej sprawdzić wiele przypadków.


```python
%%writefile test_parametrize.py
import pytest

def is_adult(age):
    if not isinstance(age, int):
        raise ValueError("Wiek musi być liczbą całkowitą.")
    return age >= 18


@pytest.mark.parametrize(
    "age, expected",
    [
        (18, True),
        (20, True),
        (17, False),
        (10, False),
        (65, True),
    ]
)
def test_is_adult(age, expected):
    assert is_adult(age) == expected


@pytest.mark.parametrize("invalid_age", ["18", 18.5, None])
def test_invalid_age_type(invalid_age):
    with pytest.raises(ValueError):
        is_adult(invalid_age)
```

```python
!pytest -q test_parametrize.py
```

## 14. Testowanie plików i danych wejściowych

Podczas pracy z plikami należy testować m.in.:

- brak pliku,
- pusty plik,
- błędny format danych,
- brak wymaganych kolumn,
- niepoprawne typy danych.

W `pytest` często wykorzystuje się `tmp_path`, czyli tymczasowy katalog dostępny tylko na czas wykonywania testu.


```python
%%writefile test_files.py
import pytest
from pathlib import Path

def read_text_file(path):
    path = Path(path)

    if not path.exists():
        raise FileNotFoundError("Plik nie istnieje.")

    content = path.read_text(encoding="utf-8")

    if not content.strip():
        raise ValueError("Plik jest pusty.")

    return content


def test_read_text_file(tmp_path):
    file_path = tmp_path / "data.txt"
    file_path.write_text("Przykładowe dane", encoding="utf-8")

    assert read_text_file(file_path) == "Przykładowe dane"


def test_missing_file(tmp_path):
    file_path = tmp_path / "missing.txt"

    with pytest.raises(FileNotFoundError):
        read_text_file(file_path)


def test_empty_file(tmp_path):
    file_path = tmp_path / "empty.txt"
    file_path.write_text("", encoding="utf-8")

    with pytest.raises(ValueError):
        read_text_file(file_path)
```

```python
!pytest -q test_files.py
```
