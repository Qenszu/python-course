# Zadanie podstawowe

Utwórz klasę `Student`, która będzie przechowywać podstawowe informacje o studencie. Następnie utwórz kilka obiektów tej klasy i zaprezentuj ich dane przy użyciu metod klasy.

Wymagania:

* zdefiniuj klasę Student,
* dodaj atrybuty: imie, nazwisko, indeks,
* zaimplementuj konstruktor __init__(), który ustawia wartości tych atrybutów podczas tworzenia obiektu,
* utwórz metodę przedstaw_sie(), która wypisuje dane studenta w czytelnej formie,
* dodaj metodę specjalną __str__(), aby obiekt można było wygodnie wyświetlić za pomocą print(),
* utwórz co najmniej 2 obiekty klasy Student i przetestuj działanie programu.

```python
class Student:
    def __init__(self, imie: str, nazwisko: str, indeks: int):
        self.imie = imie
        self.nazwisko = nazwisko
        self.indeks = indeks
    
    def przedstaw_sie(self):
        print(f"Czesc nazywam sie {self.imie} {self.nazwisko} oraz moj indeks to {self.indeks}")

    def __str__(self):
        return f"Imie: {self.imie}\nNazwisko: {self.nazwisko}\nIndex: {self.indeks}"
```

```python
s1 = Student("Czarek", "Szczepnek", 123)
s2 = Student("Kamil", "Sokulski", 125)

s1.przedstaw_sie()
print(s2.__str__())
```

# Zadanie średniozaawansowane

Utwórz klasę `KontoBankowe`, która będzie reprezentować konto użytkownika i umożliwiać wykonywanie podstawowych operacji finansowych. Program powinien kontrolować poprawność wykonywanych operacji.

Wymagania:

* zdefiniuj klasę `KontoBankowe`,
* utwórz prywatny atrybut `__saldo`, który przechowuje aktualny stan konta,
* zaimplementuj metodę `wplac(kwota)`, która zwiększa saldo po podaniu poprawnej kwoty,
* zaimplementuj metodę `wyplac(kwota)`, która zmniejsza saldo tylko wtedy, gdy na koncie są wystarczające środki,
* dodaj metodę `pokaz_saldo()`, która wyświetla aktualne saldo,
* zastosuj walidację danych, np. brak możliwości wpłaty lub wypłaty wartości ujemnej oraz brak możliwości wypłaty większej kwoty niż dostępne saldo,
* wykorzystaj `@property`, aby umożliwić bezpieczny odczyt salda bez bezpośredniego dostępu do atrybutu prywatnego.

**Wskazówki: **

W zadaniu należy wykorzystać podstawowe mechanizmy programowania obiektowego w Python, w szczególności:

* definicję klasy i konstruktora (__init__()),
* enkapsulację (atrybut prywatny __saldo),
* metody instancji operujące na danych obiektu (wplac, wyplac),
* instrukcje warunkowe (if) do walidacji danych,
* dekorator @property do kontrolowanego dostępu do atrybutów.

```python
class KontoBankowe:
    def __init__(self, wlasciciel: str, saldo_poczatkowe: float = 0.0):
        self.wlasciciel = wlasciciel
        self.__saldo = float(saldo_poczatkowe)

    def wplac(self, kwota: float):
        if kwota <= 0:
            print("Kwota wpłaty musi być dodatnia.")
            return
        self.__saldo += kwota
        print(f"Wpłacono {kwota:.2f} zł. Aktualne saldo: {self.__saldo:.2f} zł")

    def wyplac(self, kwota: float):
        if kwota <= 0:
            print("Kwota wypłaty musi być dodatnia.")
            return
        if kwota > self.__saldo:
            print("Brak wystarczających środków na koncie.")
            return
        self.__saldo -= kwota
        print(f"Wypłacono {kwota:.2f} zł. Aktualne saldo: {self.__saldo:.2f} zł")

    def pokaz_saldo(self):
        print(f"Saldo konta {self.wlasciciel}: {self.__saldo:.2f} zł")

    @property
    def saldo(self):
        return self.__saldo 
```

```python
konto = KontoBankowe("Patryk Potaczek", 100.0)
konto.pokaz_saldo()
konto.wplac(50)
konto.wyplac(30)
konto.wyplac(150)
konto.pokaz_saldo()
print(f"Saldo odczytane przez property: {konto.saldo:.2f} zł")
```

# Zadanie zaawansowane

Zaprojektuj hierarchię klas opisujących pojazdy. W rozwiązaniu należy wykorzystać mechanizmy programowania obiektowego, takie jak dziedziczenie, polimorfizm, nadpisywanie metod oraz metody klasowe i statyczne.

**Wymagania**:

* utwórz klasę bazową `Pojazd`, zawierającą podstawowe informacje o pojeździe, np. marka, model, rok,
* dodaj do klasy Pojazd metodę `opis()`, która zwraca ogólny opis obiektu,
* utwórz klasy pochodne `Samochod` i `Motocykl`, które dziedziczą po klasie `Pojazd`,
* rozszerz klasy pochodne o własne atrybuty, np. liczba drzwi dla samochodu lub typ motocykla,
* nadpisz metodę `opis()` w klasach potomnych, aby zwracała bardziej szczegółowe informacje,
* dodaj metodę klasową `utworz_domyslny()`, która tworzy przykładowy obiekt z domyślnymi danymi,
* dodaj metodę statyczną `czy_poprawny_rok(rok)`, która sprawdza, czy podany rok produkcji jest poprawny,w stosunku do obecnej daty.
* utwórz przykładowe obiekty wszystkich klas i pokaż działanie przygotowanych metod.

```python
class Pojazd:
    def __init__(self, marka: str, model: str, rok: int):
        self.marka = marka
        self.model = model
        self.rok = rok
    
    def opis(self):
        print(f"Auto marki {self.marka}, model: {self.model} z roku {self.rok}")

    @classmethod
    def utworz_domyslny(self):
        return Pojazd("BMW", "E10", 1999)
    
    @staticmethod
    def czy_poprawny_rok(rok):
        obecny_rok = 2026
        if rok > obecny_rok:
            print("Bledny rok (za duzy)")
            return False
        if rok < 1886:
            print("Pierwsze auto wytworzono dopiero w 1886")
            return False
        print("Poprawny rok")
        return True
    
class Samochod(Pojazd):
    def __init__(self, marka: str, model: str, rok: int, liczba_drzwi: int):
        super().__init__(marka, model, rok)
        self.liczba_drzwi = liczba_drzwi

    def opis(self):
        print(f"Auto marki {self.marka}, model: {self.model} z roku {self.rok} o liczbie drzwi rownej {self.liczba_drzwi}")

    @classmethod
    def utworz_domyslny(self):
        return Samochod("BMW", "E10", 1999, 3)
    

class Motocykl(Pojazd):
    def __init__(self, marka: str, model: str, rok: int, typ: str):
        super().__init__(marka, model, rok)
        self.typ = typ

    def opis(self):
        print(f"Motocykl marki {self.marka}, model {self.model} z roku {self.rok} oraz typ {self.typ}")

    @classmethod
    def utworz_domyslny(self):
        return Motocykl("MZK", "125", 1984, "szybki")

    
```

```python
p = Pojazd.utworz_domyslny()
p.opis()

s = Samochod.utworz_domyslny()
s.czy_poprawny_rok(1900)

m = Motocykl("MZK", "600", 2010, "najszybszy")
m.opis()
```

9. wykorzystaj co najmniej jeden dodatkowy mechanizm OOP, np. dziedziczenie albo polimorfizm.

* dziedziczenie (przykład):

utwórz klasę bazową `Osoba`,
klasa Uzytkownik dziedziczy po Osoba,

* lub polimorfizm (przykład):

zdefiniuj metodę `opis()` w różnych klasach (Ksiazka, Uzytkownik),
każda klasa powinna zwracać inny opis,

```python
class Ksiazka:
    def __init__(self, id: int, tytul: str, autor: str, dostepna: bool):
        self.id = id
        self.tytul = tytul
        self.autor = autor
        self.__dostepna = dostepna

    @property
    def dostepna(self):
        return self.__dostepna

    def __str__(self):
        return f'Ksiazka:\nid => {self.id}\nTytul i autor => "{self.tytul}" {self.autor}\nDostepna => {self.dostepna}'
    
    def czy_dostepna(self):
        return self.__dostepna
    
    def wypozycz(self):
        self.__dostepna = False
    
    def zwroc(self):
        self.__dostepna = True        
```

```python
class Osoba:
    def __init__(self, imie, nazwisko):
        self.imie = imie
        self.nazwisko = nazwisko
        

class Uzytkownik(Osoba):
    def __init__(self, id: int, imie: str, nazwisko: str, liczba_ksiazek: int = 0):
        super().__init__(imie, nazwisko)
        self.id = id
        self.liczba_ksiazek = liczba_ksiazek
        self.__wypozyczone_ksiazki = []

    
    def __str__(self):
        return f"Uzytkownik:\nid => {self.id}\nImie i Nazwisko => {self.imie} {self.nazwisko}\nLiczba ksiazek => {self.liczba_ksiazek}"
    
    def wypozyczone_ksiazki(self):
        for ksiazka in self.__wypozyczone_ksiazki:
            print(ksiazka)

    def wypozycz_ksiazke(self, id_ksiazki):
        self.__wypozyczone_ksiazki.append(id_ksiazki)
        self.liczba_ksiazek += 1

    def oddaj_ksiazke(self, id_ksiazki):
        self.__wypozyczone_ksiazki.remove(id_ksiazki)
        self.liczba_ksiazek -= 1
    
    def posiada_ksiazke(self, id_ksiazki):
        return id_ksiazki in self.__wypozyczone_ksiazki
```

```python
class Biblioteka:
    def __init__(self):
        self.liczba_uzytkownikow = 0
        self.uzytkownicy = {}
        self.liczba_ksiazek = 0
        self.ksiazki = {}

    def czy_dostepna(self, id):
        ksiazka = self.ksiazki.get(id, 0)
        if ksiazka == 0:
            print("Brak ksiazki w bazie")
            return False
        print(ksiazka)
        return ksiazka.czy_dostepna()

    def dodaj_ksiazke(self, tytul, autor):
        ksiazka = Ksiazka(self.liczba_ksiazek, tytul, autor, True)
        self.ksiazki[self.liczba_ksiazek] = ksiazka
        self.liczba_ksiazek += 1

    def dodaj_uzytkownika(self, imie: str, nazwisko: str):
        uzytkownik = Uzytkownik(self.liczba_uzytkownikow, imie, nazwisko, 0)
        self.uzytkownicy[self.liczba_uzytkownikow] = uzytkownik
        self.liczba_uzytkownikow += 1
        print("Dodano ", uzytkownik)

    def lista_uzytkownikow(self):
        print("--------UZYTKOWNICY--------")
        for uzytkownik in self.uzytkownicy.values():
            print(uzytkownik)
    
    def lista_ksiazek(self):
        print("----------KSIAZKI----------")
        for ksiazka in self.ksiazki.values():
            print(ksiazka)

    def wypozyczanie(self, id_uzytkownika, id_ksiazki):
        uzytkownik = self.uzytkownicy.get(id_uzytkownika, 0)
        ksiazka = self.ksiazki.get(id_ksiazki, 0)
        if uzytkownik == 0:
            print("Brak uzytkownika w bazie")
            return 
        if ksiazka == 0:
            print("Brak ksiazki w bazie")
            return
        if ksiazka.czy_dostepna():
            uzytkownik.wypozycz_ksiazke(id_ksiazki)
            ksiazka.wypozycz()
            return
        print("Ksiazka nie dostepna")
        return 
    
    def zwroc(self, id_uzytkownika, id_ksiazki):
        uzytkownik = self.uzytkownicy.get(id_uzytkownika, 0)
        ksiazka = self.ksiazki.get(id_ksiazki, 0)
        if uzytkownik == 0:
            print("Brak uzytkownika w bazie")
            return 
        if ksiazka == 0:
            print("Brak ksiazki w bazie")
            return 
        if uzytkownik.posiada_ksiazke(id_ksiazki):
            uzytkownik.oddaj_ksiazke(id_ksiazki)
            ksiazka.zwroc()
            print(f'Uzytkownik {uzytkownik.imie} zwrocil ksiazke "{ksiazka.tytul}"')
            return
        print(f'Uzytkownik {uzytkownik.imie} nie posiada ksiazki "{ksiazka.tytul}"')
```

```python
biblioteka = Biblioteka()
uzytkownicy = [("Patryk", "Nowak"), ("Kamil", "Potaczek"), ("Cezary", "Nowopolski")]
ksiazki = [("Lalka", "Boleslaw Prus"), ("Szczescie czy fart?", "Alex Rovira"), ("Obcy", "Albert Camus")]
for i, n in uzytkownicy:
    biblioteka.dodaj_uzytkownika(i, n)

for t, a in ksiazki:
    biblioteka.dodaj_ksiazke(t, a)
```

```python
biblioteka.wypozyczanie(0, 2)
biblioteka.wypozyczanie(1, 1)
biblioteka.wypozyczanie(2, 2)   # Nie mozna wypozyczyc dwa razy tej samej ksiazki
biblioteka.wypozyczanie(10, 0)  # Za duzy indeks uzytkownika
biblioteka.wypozyczanie(0, 10)  # za duzy indeks ksiazki
```

```python
biblioteka.lista_ksiazek()
biblioteka.lista_uzytkownikow()
```

```python
biblioteka.zwroc(10, 0)  # Za duzy indeks uzytkownika
biblioteka.zwroc(0, 10)  # Za duzy indeks ksiazki
biblioteka.zwroc(0, 2)
biblioteka.zwroc(0, 1)
```
